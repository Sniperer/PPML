```python
import math
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple
def fx(x: float) -> Tuple[float,float,float]:
    # 使用传入的参数 x - (1/48) * x**3 + (1/480) * x**5 - (17/80640) * x**7 + (4369/92897280) * x**9
    Taylor = 0.5 + (1/4) * x - (1/48) * x**3 + (1/480) * x**5 - (17/80640) * x**7 + (4369/92897280) * x**9
#     if(Taylor>=1):
#         Taylor=1
#     if(Taylor<=0):
#         Taylor=0
    Truth = 1 / (1 + math.exp(-x))
    Deviation = Truth - Taylor
    return (Deviation,Taylor,Truth)
```

```python
result = fx(2)
Deviation, Taylor, Truth = result
print(result)
```

```python
x_values = np.linspace(-2, 2, 400)

# 计算 y 值
taylor_values = [fx(x)[1] for x in x_values]
truth_values = [fx(x)[2] for x in x_values]
print(truth_values)
# 绘制图像
plt.figure(figsize=(10, 6))
plt.plot(x_values, taylor_values, label='Taylor Approximation', linestyle='--')
plt.plot(x_values, truth_values, label='True Function')

# 添加图例和标签
plt.title('Comparison of Taylor Approximation and True Function')
plt.xlabel('x')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

# 显示图像
plt.show()
```
